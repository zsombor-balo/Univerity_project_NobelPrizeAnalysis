1. The most Nobel of Prizes


The Nobel Prize is perhaps the world's most well known scientific award. Except for the honor, prestige and substantial prize money the recipient also gets a gold medal showing Alfred Nobel (1833 - 1896) who established the prize. Every year it's given to scientists and scholars in the categories chemistry, literature, physics, physiology or medicine, economics, and peace. Well, we're going to find out! The Nobel Foundation has made a dataset available of all prize winners from the start of the prize, in 1901, to 2016. Let's load it in and take a look.

2. So, who gets the Nobel Prize?

Just looking at the first couple of prize winners, or Nobel laureates as they are also called, we already see a celebrity: Wilhelm Conrad Röntgen, the guy who discovered X-rays. And actually, we see that all of the winners in 1901 were guys that came from Europe. But that was back in 1901, looking at all winners in the dataset, from 1901 to 2016, which sex and which country is the most commonly represented?

(For country, we will use the birth_country of the winner, as the organization_country is NaN for all shared Nobel Prizes.)

1. Which sex is the most common among all Nobel Prize winners?
2. Which birth country is the most common among all winners?


```python
'''
Which sex appears most frequently among Nobel Prize winners?
    * Look at the column sex
    * Count how many winners are Male, Female (and possibly other categories if present)
    * Identify which one appears most often.
'''
sex = nobel['sex'].value_counts()
print(sex)
print(f"{sex.idxmax()} appears most frequently among the Nobel Prize winners")
```

So men have won 836 times with women only 49 times. Cleary bias towards men

```python
'''
Which birth country appears most frequently?
    * Use the column birth_country
    * Count how many winners were born in each country
    * Identify the country with the highest count
'''
birth_country_winners = nobel['birth_country'].value_counts()
print(birth_country_winners)
print(f" {birth_country_winners.idxmax()} has the highest number of winners, as a birth country")
```

Conclusion : USA leading with 259 laureate, this country gave the most Nobel prize winners.

3. USA dominance

Not so surprising perhaps: the most common Nobel laureate between 1901 and 2016 was a man born in the United States of America. But in 1901 all the winners were European. When did the USA start to dominate the Nobel Prize charts?

```python
'''When did the United States start to dominate the Nobel Prize charts?'''
nobel['USA'] = nobel['birth_country'] == 'United States of America'
nobel['decade'] = nobel['year'] // 10 * 10 # deviding each year 
nobel['decade'].astype(int)
proportion_usa_winners = nobel.groupby('decade', as_index=False)['USA'].mean() #setting index to false to keep decade column and make things easier

proportion_usa_winners
```

Conclusion: we say that USA started to dominate the Nobel prize winners from 1930 and 1940 decade, with at least every 4th laureate came from USA, to find which specific year did the domannce begin read the next cell as a detailed complement to question #3. ⬇️⬇️⬇️

```python
year_country_counts = nobel.groupby(['year', 'birth_country']).size().reset_index(name='count')
top_countries_by_year = (
    year_country_counts
    .sort_values(['year','count'], ascending=[True, False])
    .drop_duplicates(subset=['year']))

usa_start_year = top_countries_by_year.loc[
    top_countries_by_year['birth_country'] == 'United States of America',
    'year'
].min()

print(f"The United States started the dominance in Nobel Prize winnings in {usa_start_year} (probably).")
```

4. USA dominance, visualized

A table is OK, but to see when the USA started to dominate the Nobel charts we need a plot!

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.lineplot(data=proportion_usa_winners, x='decade', y= 'USA', marker = 'o', ax =ax, color='g', errorbar=('ci', True) )
plt.title('Proportion of Nobel prize winner from the USA by decade')
plt.grid()
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
plt.xticks(rotation=45)
plt.show()
```

5. What is the gender of a typical Nobel Prize winner?

So the USA became the dominating winner of the Nobel Prize first in the 1930s and had kept the leading position ever since. But one group that was in the lead from the start, and never seems to let go, are men. Maybe it shouldn't come as a shock that there is some imbalance between how many male and female prize winners there are, but how significant is this imbalance? And is it better or worse within specific prize categories like physics, medicine, literature, etc.?

```python
#calculating the proportion of gender winners
gender_proportion = nobel['sex'].value_counts(normalize=True) #with the normalize parameter we get the proportion
nobel['female_winner'] = nobel['sex'] == 'Female'
proportion_female_winners = nobel.groupby(['decade','category'], as_index=False)['female_winner'].mean()
proportion_female_winners.tail()
```

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.lineplot(data=proportion_female_winners, x="decade", y='female_winner', hue='category', marker='o', errorbar=('ci', True), ax=ax)
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(1.0))
plt.title('Proportion of female winners by decade and category')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

You can see it on the plot that women have a higher chance of winnig a Nobel Prize for Literature & Peace with barely any chance in science.

```python
'''Who won the Prize more than once'''
more_than_once = nobel['full_name'].value_counts()
more_than_once = more_than_once[more_than_once > 1]
print(more_than_once)
```

8. How old are you when you get the prize?

The list of repeat winners contains some illustrious names! We again meet Marie Curie, who got the prize in physics for discovering radiation and in chemistry for isolating radium and polonium. John Bardeen got it twice in physics for transistors and superconductivity, Frederick Sanger got it twice in chemistry, and Linus Carl Pauling got it first in chemistry and later in peace for his work in promoting nuclear disarmament. We also learn that organizations also get the prize as both the Red Cross and the UNHCR have gotten it twice.

But how old are you generally when you get the prize?

```python
''''How old are you when you get the Nobel Prize?'''

#first we nned to convert to datetime format the 'birth_date' column
nobel['birth_date'] = pd.to_datetime(nobel['birth_date'], format='%Y-%m-%d') #notice we dont use '/' in format, data seperatyed by '- 

# Calculating the age of Nobel Prize winners
nobel['age_winner'] = nobel['year'] - nobel['birth_date'].dt.year#converting to datetime format with .dt.year extracting the year only 

#plot
sns.lmplot(data=nobel, x = 'year', y = 'age_winner', lowess=True, aspect=1.8, scatter_kws={'s' : 12.5, 'alpha':0.7}, line_kws={'lw': 2, 'color': 'red'})
plt.title('"Age of Nobel Prize Winners Over Decades')
plt.grid()
plt.plot()
```

9. Age differences between prize categories

The plot above shows us a lot! We see that people use to be around 55 when they received the price, but nowadays the average is closer to 65. But there is a large spread in the laureates' ages, and while most are 50+, some are very young.

We also see that the density of points is much high nowadays than in the early 1900s -- nowadays many more of the prizes are shared, and so there are many more winners. We also see that there was a disruption in awarded prizes around the Second World War (1939 - 1945).

Let's look at age trends within different prize categories.

```python
'''How age differs accross categories? '''
sns.lmplot(x='year', y='age_winner', data=nobel, row='category')
plt.show()
```

10. Oldest and youngest winners

More plots with lots of exciting stuff going on! We see that both winners of the chemistry, medicine, and physics prize have gotten older over time. The trend is strongest for physics: the average age used to be below 50, and now it's almost 70. Literature and economics are more stable. We also see that economics is a newer category. But peace shows an opposite trend where winners are getting younger!

In the peace category we also a winner around 2010 that seems exceptionally young. This begs the questions, who are the oldest and youngest people ever to have won a Nobel Prize?

```python
'''oldest and youngest winners'''
oldest_winners = nobel.nlargest(5, 'age_winner')
display(oldest_winners)

youngest_winners = nobel.nsmallest(5, 'age_winner')
display(youngest_winners)
```

We can see that the youngest ever laureate is Malala Yousafzai (17) from Pakistan in Peace category.

While the oldest ever winner is Leonid Hurwicz (90) from Russia in Economics category.
